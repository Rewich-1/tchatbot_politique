```python
url = "https://www2.assemblee-nationale.fr/deputes/liste/tableau"
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")
table = soup.find('table')
table_rows = table.find_all('tr')

all = {
    "adresse" : [],
    "nom" : [],
    "prenom" : [],
    "departement" : [],
    "numero_de_departement" : []
}
for tr in table_rows:
    info = []
    
    td = tr.find_all('td')
   
    links = tr.find_all('a')
    for a in tr.find_all('a', href=True):
        all["adresse"].append("https://www2.assemblee-nationale.fr"+ a['href'])
    
   
    for i in td:
        info.append(i.text.replace("\n","").replace("M. ","").replace("Mme ",""))
    #print(info)  
    if info != []:
        all["nom"].append(info[0].split(" ")[1])
        all["prenom"].append(info[0].split(" ")[0])
        all["departement"].append(info[1])
        all["numero_de_departement"].append(info[2])

    
#print(all)
```

```python
with open("depute.json", "w") as outfile:
    json.dump(all, outfile)
```
